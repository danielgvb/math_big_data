```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "twomoons.txt"

data = pd.read_csv(file_path, header=None) # read data

data.columns = ['x','y'] # rename cols

data_matrix = np.array(data) # save as array
```

# 5.4.2
-------------------

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist

import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')



# Create a range of values for k
k_range = range(1, 20)

# Initialize an empty list to 
# store the inertia values for each k
inertia_values = []

# Fit and plot the data for each k value

for k in k_range:
	kmeans = KMeans(n_clusters=k, \
					init='k-means++', random_state=42)
	y_kmeans = kmeans.fit_predict(data_matrix)
	inertia_values.append(kmeans.inertia_)

# Plot the inertia values for each k
plt.plot(k_range, inertia_values, 'bo-')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

Looks like k = 5, but the plot is cleaerly a set of two bodies shape eliptical, could use maybe dbscan or other clustering to corroborate.

```python
import networkx as nx
from sklearn.neighbors import NearestNeighbors

num_nodes = data_matrix.shape[0]

# Compute pairwise distances between all pairs of nodes
distances = np.linalg.norm(data_matrix[:, None] - data_matrix, axis=-1)

# Set the number of nearest neighbors (k)
k = 5

# Find the k-nearest neighbors for each node
knn = NearestNeighbors(n_neighbors=k+1, metric='euclidean', algorithm='auto')
knn.fit(data_matrix)
distances, indices = knn.kneighbors(data_matrix)

# Construct the k-nearest neighbor graph
G = nx.Graph()
for i in range(num_nodes):
    for j in indices[i][1:]:
        G.add_edge(i, j, weight=1)

# Plot the graph
plt.figure(figsize=(8, 6))
nx.draw(G, data_matrix, node_size = 100, node_color='green', edge_color = 'black')

plt.title('k-Nearest Neighbor Graph')
plt.show()
```

# 5.4.3 & 5.4.4
------------------------

```python
from sklearn.neighbors import kneighbors_graph

knn_graph = kneighbors_graph(data_matrix, n_neighbors=k, mode='connectivity', metric='euclidean')

# Convert the k-nearest neighbor graph to an adjacency matrix
adjacency_matrix = knn_graph.toarray()

# Degree matrix
degree_matrix = np.diag(np.sum(adjacency_matrix, axis=1))

# Laplacian
laplacian_matrix = degree_matrix - adjacency_matrix

# First 6 eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)
num_eigenvectors = 6

# Plot the first 6 eigenvectors
plt.figure(figsize=(12, 8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.scatter(data_matrix[:, 0], data_matrix[:, 1], c=eigenvectors[:, i], cmap='viridis', edgecolors='k', alpha=0.8)
    plt.title(f"Eigenvector {i+1}")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.colorbar()
plt.tight_layout()
plt.show()
```

The second eigenvector is ussualy important for spectral clustering. In this case one could see a separation of the two moon shaped clusteds.  The plot indicates a gradient of colors that divides the data into two distinct groups, with the color bar showing values transitioning from one cluster to the other. This eigenvector effectively captures the primary structure of the data.

```python
# Select the first 6 eigenvectors
eigenvectors_six = eigenvectors[:, :num_eigenvectors+1]

# Plot the scatter plots of each pair of the first 6 eigenvectors
plt.figure(figsize=(15, 15))
plot_index = 1
for i in range(num_eigenvectors):
    for j in range(i + 1, num_eigenvectors+1):
        plt.subplot(num_eigenvectors, num_eigenvectors, plot_index)
        plt.scatter(eigenvectors_six[:, i], eigenvectors_six[:, j], c=data_matrix[:, 0], cmap='viridis', edgecolors='k', alpha=0.8)
        plt.xlabel(f"Eigenvector {i+1}")
        plt.ylabel(f"Eigenvector {j+1}")
        plot_index += 1

plt.tight_layout()
plt.show()
```

From this previous plot apparently all pairs of eigenvectors from 1 to 6 seem to separate the data in tho perpendicular axis. All but the combination of eigenvectors 3 and 5. This could indicate that any eigenvector from that sample could separate the data into the two main bodies that we observe in the datacloud.

In summary, the second eigenvector is the one that better separates the data into two "distinct" clusters.

### Increase K

```python
# Spectral clustering and plot results
def spectral_clustering(data_matrix, k):
    # KNN graph
    knn_graph = kneighbors_graph(data_matrix, n_neighbors=k, mode='connectivity', metric='euclidean')

    # Adjacency matrix from graph
    adjacency_matrix = knn_graph.toarray()

    # Degree matrix
    degree_matrix = np.diag(np.sum(adjacency_matrix, axis=1))

    # Laplacian
    laplacian_matrix = degree_matrix - adjacency_matrix

    # Eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)

    # Separate using second eigenval
    second_eigenvector = eigenvectors[:, 1]
    cluster_labels = np.sign(second_eigenvector)

    # Plot points colored by cluster
    plt.figure(figsize=(8, 6))
    plt.scatter(data_matrix[:, 0], data_matrix[:, 1], c=cluster_labels, cmap='coolwarm', edgecolors='k', alpha=0.8)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Clusters based on Second Eigenvector (k={k})')
    plt.show()

for k in [5, 10, 20]:
    spectral_clustering(data_matrix, k)
```

When you increase the value of K, the graph becomes denser, so making a clustering into distinct clusters will be more difficult because each point is very conected to other points so there is no clear separation between clusters. This is seen in the values k = 10 and k = 20. Where the red cluster grows larger, leaving only a little part of the blue cluster aside.
