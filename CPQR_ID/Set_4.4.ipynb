# 4.4.0 Import Data

```python
from PIL import Image
import numpy as np

# Load the image
image_path = 'eagle.png' 
image = Image.open(image_path)

# Convert to grayscale
grayscale_image = image.convert("L")  

# Convert to a np array
data_matrix = np.array(grayscale_image)

print("Shape of data matrix:", data_matrix.shape)

print("Sample of data matrix:")
print(data_matrix[:10, :10]) 
```

## 4.4.1 Rank 10 approximation

```python
from scipy.linalg import qr
A = data_matrix
# Rank 10
k = 10

Q, R, p = qr(A, pivoting=True)
m,n = A.shape
P = np.eye(n)[:,p]

R1 = R[:k, :k]  # Upper triangular matrix for first k columns
R2 = R[:k, k:]  # R2 is the rest of the first k rows
R3 = R[k:, k:]  # Remaining rows and columns
T = np.linalg.inv(R1) @ R2  # Matrix T

I_k = np.eye(k)  # Identity matrix of size k
Z = np.hstack([I_k, T])  # Horizontal stack of I_k and T

#  Index set J
J = P[:k]  # The first k indices from the permutation

#  Compressed representation A_{:, J} and Z
C = A[:,p[:k]]

# Approximation of A
approx_A = C @ Z @ P.T

# Approximation error
approximation_error = np.linalg.norm(A - approx_A)
```

### 4.2 Plot approximation errors

```python
# list approximation errors
approximation_errors = []


for k in range(1, 101):
    
    # R1, R2, R3 matrices for decomposition
    R1 = R[:k, :k]
    R2 = R[:k, k:]
    R3 = R[k:, k:]
    
    T = np.linalg.inv(R1) @ R2
    I_k = np.eye(k)
    Z = np.hstack([I_k, T])
    C = A[:, p[:k]]
    approx_A = C @ Z @ P.T
    
    # Calculate the approximation error
    approximation_error = np.linalg.norm(A - approx_A)
    
    # Append error to the list
    approximation_errors.append(approximation_error)

# Plot of approximation errors against k
plt.plot(range(1, 101), approximation_errors)
plt.xlabel("k")
plt.ylabel("Approximation Error")
plt.title("Approximation Error vs. Rank (k)")
plt.show()
```

**A** seems to be low rank, the approximation error decreases after rank 40 and stabilizes there. This, compared to the full size of the matrix is a pretty low rank.
